```python
import busc2
from TensorFlowUtils import pt
import tensorflow as tf
from sklearn.model_selection import train_test_split
import os
import cv2
```

```python
data =    "C:\\Users\\lobo_\\Documents\\Machine Learning\\TDI\\img\\classification sets\\data\\"
numero_clases = 2
```

```python
print("Iniciando!")
searcher = busc2.Buscador(path=data, numero_clases=numero_clases)
```

```python
searcher.encuentra_conjuntos_entrenamiento_test_desde_path()
print("Imagenes cargadas!")
```

```python
x_data, y_label = searcher.X_data, searcher.y_labels  # Train Set
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,y_label,test_size=0.33, random_state=101)
```

```python
cnt = 0
path1 = "C:\\Users\\lobo_\\Documents\\Machine Learning\\TDI\\img\\classification sets\\sets\\train\\"
path2 = "C:\\Users\\lobo_\\Documents\\Machine Learning\\TDI\\img\\classification sets\\sets\\test\\"
```

```python
os.chdir(path1)
    
for x in X_train:

    image = cv2.imread(x, 1) 
    name = os.path.basename(x)
    #cv2.imshow(name, image) 

    #si hay 0 o 1 en la dir de la img sabremos a que clase pertenece
    # y asi podemos designar a cual folder copiarla
    if 'class0' in x:
        os.chdir(os.path.join(path1,'0')) #nos cambiamos a este folder
        cv2.imwrite(name, image)
    if 'class1' in x:
        os.chdir(os.path.join(path1,'1'))
        cv2.imwrite(name, image)
            
```
